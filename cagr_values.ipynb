```python
# Compute 5 year CAGR for Marine Premiums for relevant countries.
```

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("C:\\Users\\gupta\\OneDrive\\Desktop\\excel\\insurance_csv.csv" )
df
```

```python
marine_premiums = df[(df['lob'] == 'marine') & (df['variable_name'] == 'Premiums')]
marine_premiums
```

```python
# marine_premiums['Country']
```

```python
# Define the function to calculate CAGR
def calculate_cagr(beginning_value, ending_value, num_years):
    if beginning_value > 0 and ending_value > 0:
        return (ending_value / beginning_value) ** (1 / num_years) - 1
    else:
        return None
```

```python
# Calculate 5-year CAGR for each country
years = [2001, 2006, 2011]  # Define the 5-year periods
cagr_results = []
for country in marine_premiums['Country'].unique():
    country_data = marine_premiums[marine_premiums['Country'] == country]
    # print([marine_premiums['Country'] == country])
    for year in years:
        if str(year) in country_data.columns and str(year + 4) in country_data.columns:
            beginning_value = country_data[str(year)].values[0]
            # print(country_data[str(year)].values[0])
            ending_value = country_data[str(year + 4)].values[0]
            # print(country_data[str(year + 5)].values[0])
            if pd.notna(beginning_value) and pd.notna(ending_value):
                cagr = calculate_cagr(beginning_value, ending_value, 5)
                if cagr is not None:
                    cagr_results.append({'Country': country, 'Start Year': year, 'End Year': year + 4, 'CAGR': cagr})
```

```python
# Convert the results to a DataFrame
cagr_df = pd.DataFrame(cagr_results)
```
