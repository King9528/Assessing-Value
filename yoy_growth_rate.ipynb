```python
# Compare YoY growth rates for global aviation premiums for actual value based countries vs assumption based countries at the aggregate level.
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("C:\\Users\\gupta\\OneDrive\\Desktop\\excel\\insurance_csv.csv" )
df
```

```python
aviation_premiums = df[(df['lob'] == 'aviation') & (df['variable_name'] == 'Premiums')]
aviation_premiums
```

```python
actual_value_countries = aviation_premiums[aviation_premiums['comments'].isna()]
assumption_based_countries = aviation_premiums[aviation_premiums['comments'].notna()]
```

```python
print(actual_value_countries)
print(assumption_based_countries)
```

```python
def calculate_yoy_growth(df):
    yearly_columns = df.columns[6:]  # Columns representing the years
    # df = df.copy()
    df[yearly_columns] = df[yearly_columns].replace(0, np.nan)  # Replace 0 with NaN to avoid division by zero issues
    aggregate_premiums = df[yearly_columns].sum()
    yoy_growth = aggregate_premiums.pct_change() * 100  # Calculate YoY growth rates in percentage
    return yoy_growth
```

```python
# Calculate the YoY growth rates for each group
actual_value_yoy_growth = calculate_yoy_growth(actual_value_countries)
assumption_based_yoy_growth = calculate_yoy_growth(assumption_based_countries)
```

```python
# Display the results
print("Actual Value-Based Countries YoY Growth Rates:")
print(actual_value_yoy_growth)
```

```python

print("\nAssumption-Based Countries YoY Growth Rates:")
print(assumption_based_yoy_growth)
```
